```python
# Asumsikan nama kolom teks adalah 'full_text', ganti jika perlu
NAMA_KOLOM_TEKS = 'full_text'

# Menghapus baris dimana kolom teksnya kosong (NaN)
df.dropna(subset=[NAMA_KOLOM_TEKS], inplace=True)

# Menghapus baris yang merupakan duplikat sempurna
df.drop_duplicates(inplace=True)

# Menghapus duplikat berdasarkan teks cuitan saja
df.drop_duplicates(subset=[NAMA_KOLOM_TEKS], keep='first', inplace=True)

print(f"Jumlah data setelah dibersihkan dari data kosong dan duplikat: {len(df)}")
```

```python
import re

def clean_text(text):
    text = str(text).lower()  # Mengubah ke huruf kecil dan memastikan tipe data string
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE) # Hapus URL

    # --- BARIS YANG DIPERBAIKI ---
    text = re.sub(r'@\w+|#\w+|rt', '', text) # Hapus mention, hashtag, dan rt

    text = re.sub(r'[^a-zA-Z\s]', '', text) # Hapus tanda baca dan karakter spesial
    text = re.sub(r'\d+', '', text) # Hapus angka
    text = " ".join(text.split()) # Hapus spasi berlebih di tengah teks
    text = text.strip() # Hapus spasi di awal dan akhir
    return text

df['cleaned_text'] = df[NAMA_KOLOM_TEKS].apply(clean_text)
```

```python
output_filename = '/content/tweets-data/purbaya_prepro.csv'
df.to_csv(output_filename, index=False, encoding='utf-8-sig')
```

```python
import pandas as pd
import re

normalization_dict = {
    'yg': 'yang', 'jg': 'juga', 'ga': 'tidak', 'gak': 'tidak', 'gaada': 'tidak ada',
    'ngga': 'tidak', 'nggak': 'tidak', 'klo': 'kalau', 'kalo': 'kalau', 'utk': 'untuk',
    'dlm': 'dalam', 'bgt': 'banget', 'dgn': 'dengan', 'dr': 'dari', 'sdh': 'sudah',
    'blm': 'belum', 'lg': 'lagi', 'sm': 'sama', 'tp': 'tapi', 'sy': 'saya', 'tdk': 'tidak',
    'jgn': 'jangan', 'org': 'orang', 'krn': 'karena', 'emg': 'memang', 'dg': 'dengan',
    'btw': 'by the way', 'otw': 'on the way', 'mantap': 'bagus', 'keren': 'bagus',
    'iyaa': 'iya', 'okee': 'oke', 'siapp': 'siap', 'gpp': 'tidak apa-apa'
    # Tambahkan kata lainnya di sini...
}
```

```python
def normalize_text(text):
    # --- 2. Menormalkan kata yang hurufnya berulang ---
    # Contoh: "iyaaaa" -> "iyaa" (dibatasi 2 perulangan untuk kata seperti "maaf")
    text = re.sub(r'(.)\1+', r'\1\1', text)

    words = text.split()
    normalized_words = []
    for word in words:
        # --- 3. Normalisasi kata typo, singkatan, dan gaul menggunakan kamus ---
        if word in normalization_dict:
            word = normalization_dict[word]
        normalized_words.append(word)

    text = " ".join(normalized_words)

    # --- 4. Menghapus kata yang kurang dari 3 huruf ---
    words = text.split()
    long_words = [word for word in words if len(word) >= 3]
    text = " ".join(long_words)

    return text

if 'cleaned_text' in df.columns:
    df['normalized_text'] = df['cleaned_text'].apply(normalize_text)

    print("✅ Sukses! Contoh hasil normalisasi:")
    # Tampilkan perbandingan kolom-kolomnya untuk melihat perubahan
    print(df[['cleaned_text', 'normalized_text']].head(10))
else:
    print("❌ Error: Kolom 'cleaned_text' tidak ditemukan. Pastikan Anda sudah menjalankan kode preprocessing sebelumnya.")
```

```python
# Instalasi library wordcloud
!pip install wordcloud -q

# Impor library yang diperlukan
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Muat file CSV yang sudah diproses
file_path = '/content/tweets-data/purbaya_normalized.csv'
df = pd.read_csv(file_path)

# Pastikan kolom teks tidak memiliki nilai kosong (NaN)
df.dropna(subset=['normalized_text'], inplace=True)

print(f"Data berhasil dimuat. Jumlah baris: {len(df)}")
```

```python
# --- UNIGRAMS (1-GRAM) ---

# Membuat unigram dari kolom 'normalized_text'
unigram_vectorizer = CountVectorizer(ngram_range=(1, 1))
unigram_matrix = unigram_vectorizer.fit_transform(df['normalized_text'])

# Periksa apakah ada kata yang ditemukan
if len(unigram_vectorizer.vocabulary_) > 0:
    # Hitung frekuensi setiap kata
    frequencies = unigram_matrix.sum(axis=0).tolist()[0]
    feature_names = unigram_vectorizer.get_feature_names_out()
    unigram_freq = dict(zip(feature_names, frequencies))

    # Buat dan tampilkan Word Cloud
    wc = WordCloud(width=800, height=400, background_color='white', collocations=False)
    wc.generate_from_frequencies(unigram_freq)

    plt.figure(figsize=(10, 5))
    plt.imshow(wc, interpolation='bilinear')
    plt.title('Word Cloud Unigrams (1-gram)', fontsize=16)
    plt.axis('off')
    plt.show()
else:
    print("Tidak dapat membuat Word Cloud Unigram (tidak ada kata yang valid ditemukan).")
```

```python
# --- BIGRAMS (2-GRAM) ---

# Membuat bigram dari kolom 'normalized_text'
bigram_vectorizer = CountVectorizer(ngram_range=(2, 2))
bigram_matrix = bigram_vectorizer.fit_transform(df['normalized_text'])

# Periksa apakah ada pasangan kata yang ditemukan
if len(bigram_vectorizer.vocabulary_) > 0:
    # Hitung frekuensi setiap pasangan kata
    frequencies = bigram_matrix.sum(axis=0).tolist()[0]
    feature_names = bigram_vectorizer.get_feature_names_out()
    bigram_freq = dict(zip(feature_names, frequencies))

    # Buat dan tampilkan Word Cloud
    wc = WordCloud(width=800, height=400, background_color='white', collocations=False)
    wc.generate_from_frequencies(bigram_freq)

    plt.figure(figsize=(10, 5))
    plt.imshow(wc, interpolation='bilinear')
    plt.title('Word Cloud Bigrams (2-gram)', fontsize=16)
    plt.axis('off')
    plt.show()
else:
    print("Tidak dapat membuat Word Cloud Bigram (kemungkinan teks terlalu pendek).")
```

```python
# --- TRIGRAMS (3-GRAM) ---

# Membuat trigram dari kolom 'normalized_text'
trigram_vectorizer = CountVectorizer(ngram_range=(3, 3))
trigram_matrix = trigram_vectorizer.fit_transform(df['normalized_text'])

# Periksa apakah ada frasa tiga kata yang ditemukan
if len(trigram_vectorizer.vocabulary_) > 0:
    # Hitung frekuensi setiap frasa
    frequencies = trigram_matrix.sum(axis=0).tolist()[0]
    feature_names = trigram_vectorizer.get_feature_names_out()
    trigram_freq = dict(zip(feature_names, frequencies))

    # Buat dan tampilkan Word Cloud
    wc = WordCloud(width=800, height=400, background_color='white', collocations=False)
    wc.generate_from_frequencies(trigram_freq)

    plt.figure(figsize=(10, 5))
    plt.imshow(wc, interpolation='bilinear')
    plt.title('Word Cloud Trigrams (3-gram)', fontsize=16)
    plt.axis('off')
    plt.show()
else:
    print("Tidak dapat membuat Word Cloud Trigram (kemungkinan teks terlalu pendek).")
```
